```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
new=pd.read_csv('../Case-Study/Dataset/startup_funding.csv')
df=new.copy()
```

```python
df.info()
```

```python
df['AmountInUSD'].fillna(0,inplace=True)
```

```python
df2=df.copy()
df2['AmountInUSD']=df2['AmountInUSD'].str.replace(',','')
df2.AmountInUSD.fillna(0,inplace=True)
df2['AmountInUSD']=df2['AmountInUSD'].astype(int)
final=df2.groupby('IndustryVertical')["AmountInUSD"].sum()
final.sort_values()
```

```python
X=final.sort_values()[-1:-6:-1].index
Y=final.sort_values()[-1:-6:-1].values
```

```python
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
explode=[0,1,1,2]
plt.pie(Y,labels=X,autopct='%.2f%%',counterclock=True)
plt.show()
```
